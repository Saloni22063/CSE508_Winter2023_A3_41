```python
import numpy as np
import networkx as nx
```

```python
Directed_g = nx.read_edgelist(r'/content/drive/MyDrive/email-Eu-core.txt', create_using=nx.DiGraph)
```

PageRank Score for each node

```python
# Calculate the PageRank score for each node
pagerank_scores = nx.pagerank(Directed_g)
pagerank_scores
```

```python
# Sort the pages based on their PageRank score from high to low
sorted_pages = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)

# Print the PageRank score for each node
for node, score in sorted_pages:
  print(f"Node {node} has a PageRank score of {score:.4f}")
```

Authority and Hub score for each node

```python
# Compute the HITS algorithm for the graph
hub_scores, authority_scores= nx.hits(Directed_g, max_iter= 100, normalized = True)
```

Hub score for each node

```python
# Sort the hub score high to low
sorted_hubs = dict(sorted(hub_scores.items(), key=lambda item: item[1], reverse=True))
# Print the hub score for each node
for node, score in sorted_hubs.items():
    print(f"Node {node} has a Hub score of {score:.6f}")
```

Authority score for each node

```python
# Sort the authority score high to low
sorted_authority = dict(sorted(authority_scores.items(), key=lambda item: item[1], reverse=True))
# Print the authority score for each node
for node, score in sorted_authority.items():
    print(f"Node {node} has a authority score of {score:.6f}")
```

Comparing the results obtained from both the algorithms in parts 1 and 2 based on the node
scores.

From the results above, the pagerank and authorities work upon incoming nodes
but still shows different nodes as important because pagerank takes nodes with
maximum in-degree i.e. for node 1 whereas, authority score is calculated
using maximum incoming nodes from the hubs, so that doesn’t includes total
incoming nodes.
In Hubs, the Outgoing nodes are referred and thus, 160 has maximum
OutDegree nodes
Other than that we can observe that node 160 has miximum hub score and authority score which shows that it is a highly connected and influential node in the network. This suggests that it is both central and authoritative, and plays a key role in the network.
